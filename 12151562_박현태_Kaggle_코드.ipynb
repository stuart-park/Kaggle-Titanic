# 2. Data Preprocessing & Augmentation

### a) Data Preprocessing

#### 1) PassengerId

```python
train_df=train_df.drop(['PassengerId'], axis=1)
test_df=test_df.drop(['PassengerId'], axis=1)
```

#### 2) Cabin

```python
train_df=train_df.drop(['Cabin'], axis=1)
test_df=test_df.drop(['Cabin'], axis=1)
```

#### 3) Ticket

```python
train_df=train_df.drop(['Ticket'], axis=1)
test_df=test_df.drop(['Ticket'], axis=1)
```

```python
ax=sns.countplot(x="Embarked", data=train_df)
```

```python
combine=[train_df, test_df]
for data in combine:
    data['Embarked']=data['Embarked'].fillna('S')
```

```python
embarked_mapping={"S":0, "C":1,"Q":2}
train_df['Embarked']=train_df['Embarked'].map(embarked_mapping)
test_df['Embarked']=test_df['Embarked'].map(embarked_mapping)
```

#### 7) Fare

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)
test_df.describe()
```

```python
fare_band=[]
fare_band=pd.qcut(train_df['Fare'], 5)
fare_band
```

```python
combine=[train_df, test_df]
for data in combine:
    data.loc[data['Fare']<=7.854, 'Fare']=0
    data.loc[(data['Fare']<=10.5) & (data['Fare']>7.854), 'Fare']=1
    data.loc[(data['Fare']<=21.679) & (data['Fare']>10.5), 'Fare']=2
    data.loc[(data['Fare']<=39.688) & (data['Fare']>21.679), 'Fare']=3
    data.loc[data['Fare']>39.688, 'Fare']=4
    data['Fare']=data['Fare'].astype(int)
```

#### 10) Age

```python
guess_ages=np.zeros((2,3))
combine=[train_df, test_df]
for data in combine:
    for i in range(0,2):
        for j in range(0,3):
            temp_df=data[(data['Sex']==i)&(data['Pclass']==j+1)]['Age'].dropna()
            guess_ages[i, j]=temp_df.median()
```

```python
guess_ages
```

```python
combine=[train_df, test_df]
for data in combine:
    for i in range(0, 2):
        for j in range(0,3):
            data.loc[(data['Sex']==i)&(data['Pclass']==j+1)&(data['Age'].isnull()), 'Age']=guess_ages[i,j]
    data['Age']=data['Age'].astype(int)
    
train_df.head()
```

```python
train_df.describe()
```

```python
test_df.describe()
```

```python
age_band=[]
age_band=pd.cut(train_df['Age'], 8)
age_band
```

```python
combine=[train_df, test_df]
for data in combine:
    data.loc[data['Age']<=10.0, 'Age']=0
    data.loc[(data['Age']>10.0)&(data['Age']<=20.0), 'Age']=1
    data.loc[(data['Age']>20.0)&(data['Age']<=30.0), 'Age']=2
    data.loc[(data['Age']>30.0)&(data['Age']<=40.0), 'Age']=3
    data.loc[(data['Age']>40.0)&(data['Age']<=50.0), 'Age']=4
    data.loc[(data['Age']>50.0)&(data['Age']<=60.0), 'Age']=5
    data.loc[(data['Age']>60.0)&(data['Age']<=70.0), 'Age']=6
    data.loc[data['Age']>70, 'Age']=7
    data['Age']=data['Age'].astype(int)
```

#### 11) Name

```python
combine=[train_df, test_df]
for data in combine:
    data['Title'] = data.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
combine=[train_df, test_df]
for data in combine:
    data['Title'] = data['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    data['Title'] = data['Title'].replace('Mlle', 'Miss')
    data['Title'] = data['Title'].replace('Ms', 'Miss')
    data['Title'] = data['Title'].replace('Mme', 'Mrs')
```

```python
ax=sns.countplot(x="Title", hue="Survived", data=train_df)
```

```python
title_mapping={"Mr":0,"Mrs":1, "Miss":2, "Master":3, "Rare":4}
train_df['Title']=train_df['Title'].map(title_mapping)
test_df['Title']=test_df['Title'].map(title_mapping)
```

```python
train_df=train_df.drop(['Name'], axis=1)
test_df=test_df.drop(['Name'], axis=1)
```

```python
train_df.head()
```

```python
test_df.head()
```

### b) Data Augmentation

```python
combine=[train_df, test_df]
for data in combine:
    data['Pclass*Sex']=data.Pclass * data.Sex
    data['Pclass*Age']=data.Pclass*data.Age
    data['Pclass*Fare']=data.Pclass*data.Fare
    data['Pclass*Embarked']=data.Pclass*data.Embarked
    data['Pclass*FamilySize']=data.Pclass*data.FamilySize
    data['Pclass*Title']=data.Pclass*data.Title
    data['Sex*Age']=data.Sex*data.Age
    data['Sex*Fare']=data.Sex*data.Fare
    data['Sex*Embarked']=data.Sex*data.Embarked
    data['Sex*FamilySize']=data.Sex*data.FamilySize
    data['Sex*Title']=data.Sex*data.Title
    data['Age*Fare']=data.Age*data.Fare
    data['Age*Embarked']=data.Age*data.Embarked
    data['Age*FamilySize']=data.Age*data.FamilySize
    data['Age*Title']=data.Age*data.Title
    data['Fare*Embarked']=data.Fare*data.Embarked
    data['Fare*FamilySize']=data.Fare*data.FamilySize
    data['Fare*Title']=data.Fare*data.Title
    data['Embarked*FamilySize']=data.Embarked*data.FamilySize
    data['Embarked*Title']=data.Embarked*data.Title
    data['FamilySize*Title']=data.FamilySize*data.Title
```

# 3. 모델링과 학습

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train=train_df.drop('Survived', axis=1)
y_train=train_df['Survived']
x_test=test_df.copy()
x_train.shape, y_train.shape, x_test.shape
```

#### 1) LogisticRegression

```python
logreg=LogisticRegression()
logreg.fit(x_train, y_train)
y_pred_LR=logreg.predict(x_test)
acc=logreg.score(x_train, y_train)
acc
#score : 0.77033
```

#### 2) Support Vector Machine

```python
svc=SVC()
svc.fit(x_train, y_train)
y_pred_SVM=svc.predict(x_test)
acc=svc.score(x_train, y_train)
acc
#score : 0.78468
```

#### 3) KNN

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred_KNN=knn.predict(x_test)
acc_knn=knn.score(x_train, y_train)
acc_knn
#score : 0.77990
```

#### 4)Decision Tree

```python
decision_tree=DecisionTreeClassifier()
decision_tree.fit(x_train, y_train)
y_pred_DT=decision_tree.predict(x_test)
acc_decision_tree=decision_tree.score(x_train, y_train)
acc_decision_tree
#score : 0.75598
```

#### 5) Random Forest

```python
random_forest=RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
y_pred_RF=random_forest.predict(x_test)
acc_random_forest=random_forest.score(x_train, y_train)
acc_random_forest
#socre : 0.77511
```

#### 6)AdaBoost

```python
ada_boost=AdaBoostClassifier(n_estimators=100)
ada_boost.fit(x_train , y_train)
y_pred_ab=ada_boost.predict(x_test)
acc_ada_boost=ada_boost.score(x_train, y_train)
acc_ada_boost
#score : 0.78947
```

#### 7) Bagging

```python
bagging=BaggingClassifier(n_estimators=100)
bagging.fit(x_train, y_train)
y_pred_bag=bagging.predict(x_test)
acc_bagging=bagging.score(x_train, y_train)
acc_bagging
#score : 0.77033
```
